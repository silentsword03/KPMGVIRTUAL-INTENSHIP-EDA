```python
#check for duplicated
```

```python
#create a new column age. 
```

```python
now = pd.to_datetime('now')
CustomerDemographic['Age'] = (now - CustomerDemographic['DOB']).astype('<m8[Y]').sort_values(ascending=False)
```

```python
CustomerDemographic.head(5)
```

```python
CustomerDemographic.columns
```

```python
CustomerDemographic.Age.max()
```

```python
CustomerDemographic = CustomerDemographic[CustomerDemographic.Age <=100]
```

```python
CustomerDemographic.Age.max() #Dropped all the list of customers whose age is greater than 100. 
```

```python
CustomerDemographic.deceased_indicator.unique() #Lets drop the people from the list who are deceased for more accurate analysis in future.
```

```python
CustomerDemographic = CustomerDemographic[CustomerDemographic.deceased_indicator == 'N']
```

```python
CustomerDemographic.head(5)
```
